- Does a 10% discount lead to more sales than no discount?

Formulate and test a Hypothesis for the above questions 


```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("sample superstore.xls")
df.columns
```

```python
def Hypothesis_test(h0, h1, p_value, alpha = 0.05):
    if p_value >= alpha:
        print(f"Fail to reject h0 -> {h0}")
    elif p_value < alpha:
        print(f"Reject h0 -> {h1}")
```

```python
from scipy.stats import ttest_1samp
population_mean = df["Profit"].mean()
population_mean
```

```python
sample = df["Profit"].sample(n=100)
sample.mean()
```

```python
# right tailed test
t_stat, p_value = ttest_1samp(sample,popmean=50,alternative="greater")
print(f"t_stat{t_stat:.2f}, p_value{p_value:.4f}")
```

```python
h1 = "The average profit per order > 50 dollars"
h0 = "The average profit <= 50 dollars"
print(f"sample mean = {sample.mean()} ,population mean = {population_mean}",)
print(f't_stat = {t_stat} , p_val_right = {p_value}')
Hypothesis_test(h0, h1, p_value)
```

```python
# Is there a difference in average sales between East and West regions?
from scipy.stats import ttest_ind

E = df.loc[df['Region'] == 'East', "Sales"]
w = df.loc[df['Region'] == 'West', 'Sales']

t_stat, p_value = ttest_ind(E, w)
print(f"t-statistic = {t_stat:.4f}, p-value = {p_value:.4f}")

```

### **Tasks**

* Is the average passenger age different from 30 years?

* Is survival independent of passenger class?

**Formulate and test a hypothesis for each of the above questions.**


```python
from scipy.stats import ttest_1samp
import pandas as pd
```

```python
data = pd.read_csv("train.csv")
data.columns
```

```python
data.dropna(inplace=True)
```

```python
h1 = "age != 30"
h0 = "age = 30"
sample = data['Age'].sample(n=100)
sample.mean()
```

```python
t_stat, p_value = ttest_1samp(sample, popmean=30)
print(f"t_stat {t_stat:.3f}, P_value {p_value:3f}")
```

```python
print(f"T_test {t_stat:.3f}, P_value {p_value:3f}")
Hypothesis_test(h0, h1, p_value)
```

```python
# Did males and females pay different average fares?
from scipy.stats import ttest_ind

Males = data.loc[data['Sex'] == "male", "Fare"]
Females = data.loc[data['Sex'] == "female", "Fare"]
h1 = "males average fares != female average fares"
h0 = "males average fares = female average fares"

```

```python
t_stat, p_value = ttest_ind(Males, Females)
print(f"male_averages {Males.mean()}, Female_average {Females.mean()}")
print(f"T_test {t_stat:.3f}, P_value {p_value: .4f}")
Hypothesis_test(h0, h1, p_value)
```

```python
alpha=0.005
if p_value<alpha:
    print('reject H0:passengers do not differ in mean age')
else:
    print('Fail to reject H0: passangers in different classes differ in mean age')
```

```python
# Do passengers in different travel classes (1st, 2nd, 3rd) differ in their mean age?
from scipy.stats import f_oneway
t_class = data.groupby('Pclass')['Age'].apply(list)
f_stat, p_value = f_oneway(*t_class)
print(f"F_stat{f_stat :2f}, P_value{p_value:.3f}")
h1 = "Atleast one travel class differ in the mean age"
h0= "The travel class do not differ in the mean age"
Hypothesis_test(h0, h1, p_value)
```
